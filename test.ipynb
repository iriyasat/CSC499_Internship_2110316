```python
def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"


students = {
    "Ibrahim": 91,
    "Sara": 74,
    "Omar": 55,
    "Mia": 88,
    "Tariq": 63
}

with open("results.txt", "w") as file:
    for name, score in students.items():
        grade = get_grade(score)

        line = f"{name:<8} | {score} | {grade}\n"

        file.write(line)

print("Results written to results.txt\n")


try:
    with open("results.txt", "r") as file:
        contents = file.read()

        print("File contents:")
        print(contents)

except FileNotFoundError:
    print("Error: results.txt was not found.")
```
